## Import Semua Packages/Library yang Digunakan

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

#### Customer

```python
file_path_customers = 'data/customers_dataset.csv'
customers_df = pd.read_csv(file_path_customers)
customers_df.head()
```

#### Geolocation

```python
file_path_geo = 'data/geolocation_dataset.csv'
geolocation_df = pd.read_csv(file_path_geo)
geolocation_df.head()
```

#### Order Items

```python
file_path_order_items = 'data/order_items_dataset.csv'
order_items_df = pd.read_csv(file_path_order_items)
order_items_df.head()
```

#### Order Payments

```python
file_path_order_payments = 'data/order_payments_dataset.csv'
order_payments_df = pd.read_csv(file_path_order_payments)
order_payments_df.head()
```

#### Order Reviews

```python
file_path_order_reviews = 'data/order_reviews_dataset.csv'
order_reviews_df = pd.read_csv(file_path_order_reviews)
order_reviews_df.head()
```

#### Orders

```python
file_path_orders = 'data/orders_dataset.csv'
orders_df = pd.read_csv(file_path_orders)
orders_df.head()
```

#### Product Category Name

```python
file_path_product_category_name_translation = 'data/product_category_name_translation.csv'
product_category_name_translation_df = pd.read_csv(file_path_product_category_name_translation)
product_category_name_translation_df.head()
```

#### Products

```python
file_path_products = 'data/products_dataset.csv'
products_df = pd.read_csv(file_path_products)
products_df.head()
```

#### Sellers

```python
file_path_sellers = 'data/sellers_dataset.csv'
sellers_df = pd.read_csv(file_path_sellers)
sellers_df.head()
```

#### Merging

```python
new_orders_df = pd.merge(
    left=orders_df,
    right=order_items_df,
    how="inner",
    left_on="order_id",
    right_on="order_id"
)

new_orders_df.head()
```

```python
new_orders_product_df = pd.merge(
    left=new_orders_df,
    right=products_df,
    how="inner",
    left_on="product_id",
    right_on="product_id"
)

new_orders_product_df.head()
```

```python
new_orders_product_name_df = pd.merge(
    left=new_orders_product_df,
    right=product_category_name_translation_df,
    how="inner",
    left_on="product_category_name",
    right_on="product_category_name"
)

new_orders_product_name_df.head()
```

```python
analysis_df = pd.merge(
    left=new_orders_product_name_df,
    right=customers_df,
    how="inner",
    left_on="customer_id",
    right_on="customer_id"
)

analysis_df.head()
```

```python
selected_columns_df = analysis_df[['order_id','customer_id','product_id','order_delivered_customer_date', 'product_category_name_english', 'customer_city']]

selected_columns_df.head()
```

### Assessing Data

#### Merged Dataframe

```python
selected_columns_df.info()
```

```python
selected_columns_df.isna().sum()
```

```python
print("Jumlah duplikasi: ", selected_columns_df.duplicated().sum())
```

### Cleaning Data

#### Merged Dataframe

```python
selected_columns_df = selected_columns_df.dropna(subset=['order_delivered_customer_date'])
```

```python
selected_columns_df.isna().sum()
```

```python
selected_columns_df["order_delivered_customer_date"] = pd.to_datetime(selected_columns_df["order_delivered_customer_date"])
selected_columns_df.info()
```

```python
selected_columns_df.drop_duplicates(inplace=True)
```

```python
print("Jumlah duplikasi: ", selected_columns_df.duplicated().sum())
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
sum_order_items_df = selected_columns_df.groupby(by="product_category_name_english").order_id.nunique().sort_values(ascending=False).reset_index()
sum_order_items_df = sum_order_items_df.rename(columns={
    'product_category_name_english': 'product_name',
    'order_id': 'order_sum'
})
sum_order_items_df.head(15)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))
 
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
 
sns.barplot(x="order_sum", y="product_name", hue="order_sum", hue_order=sum_order_items_df['order_sum'].head(5), data=sum_order_items_df.head(5), palette=colors, ax=ax[0], legend=False)
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Best Performing Product", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)
 
sns.barplot(x="order_sum", y="product_name", hue="order_sum", data=sum_order_items_df.sort_values(by="order_sum", ascending=True).head(5), palette=colors, ax=ax[1], legend=False)
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Worst Performing Product", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)
 
plt.suptitle("Best and Worst Performing Product by Number of Sales", fontsize=20)
plt.show()
```

### Pertanyaan 2:

```python
sum_order_cities_df = selected_columns_df.groupby(by="customer_city").order_id.nunique().sort_values(ascending=False).reset_index()
sum_order_cities_df = sum_order_cities_df.rename(columns={
    'order_id': 'order_sum'
})
sum_order_cities_df.head(15)
```

```python
plt.figure(figsize=(10, 5))
 
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
 
sns.barplot(x="order_sum", y="customer_city", hue="order_sum", hue_order=sum_order_cities_df['order_sum'].head(5), data=sum_order_cities_df.head(5), palette=colors, legend=False)
plt.ylabel(None)
plt.xlabel(None)
plt.title("The City That Buys the Most", loc="center", fontsize=20)
plt.tick_params(axis ='y', labelsize=12)
 
plt.show()
```

## Conclusion

### Conclusion Pertanyaan 1

Kategori produk yang paling banyak terjual adalah "bed_bath_table", "health_beauty", "sports_leisure", "computers_accessories", dan "furniture_decor". Produk-produk dalam kategori ini kemungkinan memiliki permintaan yang tinggi di pasar karena merupakan kebutuhan rumah tangga, adanya tren kesehatan dan kecantikan, minat dalam olahraga dan rekreasi, serta kebutuhan akan aksesori komputer dan dekorasi rumah. Seller dapat mempertimbangkan untuk meningkatkan stok, memperluas variasi produk, dan meningkatkan strategi pemasaran untuk mempertahankan dan meningkatkan penjualan.

Kategori produk yang paling sedikit terjual adalah "security_and_services", "fashion_childrens_clothes", "cds_dvds_musicals", "la_cuisine", dan "arts_and_craftmanship". Ini bisa disebabkan oleh berbagai faktor seperti kurangnya minat pasar, persaingan yang ketat, atau produk yang tidak sesuai dengan tren saat ini. Seller dapat melakukan analisis lebih lanjut untuk memahami alasan di balik rendahnya penjualan. Mungkin perlu dilakukan penyesuaian strategi pemasaran, peningkatan kualitas produk, atau bahkan mempertimbangkan untuk mengurangi atau menghentikan penawaran produk tersebut jika tidak menguntungkan. 

### Conclusion Pertanyaan 2

Kategori produk yang paling banyak terjual berasal dari kota-kota berikut: "Sao Paulo", "Rio de Janeiro", "Belo Horizonte", "Brasilia", dan "Curitiba". Kota-kota ini memiliki aktivitas e-commerce yang tinggi, menunjukkan bahwa penduduk di kota-kota ini cenderung lebih aktif dalam melakukan pembelian online. Ini bisa disebabkan oleh berbagai faktor seperti tingkat urbanisasi, akses internet yang baik, dan daya beli yang tinggi.

Seller dapat mempertimbangkan untuk meningkatkan upaya pemasaran dan promosi di kota-kota ini untuk memaksimalkan penjualan. Misalnya, kampanye iklan yang lebih intensif atau penawaran khusus untuk pelanggan di kota-kota tersebut. Mengingat tingginya volume pembelian dari kota-kota ini, perusahaan dapat mengoptimalkan logistik dan pengiriman untuk memastikan produk sampai ke pelanggan dengan cepat dan efisien. Perusahaan dapat membangun kemitraan dengan bisnis lokal di kota-kota ini untuk meningkatkan kehadiran merek dan memperluas jaringan distribusi.
